### <span style='font-family:"Times New Roman"'>Loading Required Modules:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
plt.style.use('ggplot')
```

### <span style='font-family:"Times New Roman"'>Implemented Functions:  

```python
def guassX(N,Me,Var):
    dimension = len(Me)
    a = np.array(Me)
    p = np.random.randn(dimension,N)
    h = np.tile(a, (2, N//2))
    x = math.sqrt(Var) * p + h
    return x
```

```python
M1 = float(input("Enter x coordinate of your mean (In this problem = 0) :"))
M2 = float(input("Enter y coordinate of your mean (In this problem = 0) :"))
Var = float(input("Enter variance of data set (In this problem = 1) :"))
N = int(input("Number of sample data (In this problem = 500):"))
```

```python
Mean = [M1, M2]
X = guassX(N, Mean, Var)
df = pd.DataFrame(X).T  # transposed for easier plotting
df = df[(df[0] * df[1]) >= 0]  # slice df to get 1st and 3rd quadrant values
Y = guassX(N, Mean, Var)
dg = pd.DataFrame(Y).T  # transposed for easier plotting
dg = dg[(dg[0] * dg[1]) <= 0]  # slice df to get 1st and 3rd quadrant values
```

```python
plt.figure(figsize=(8,8))
fig1 = plt.figure(1)
ax = fig1.gca()
ax = df.plot.scatter(0,1, marker='x', c='g',label = "target : -1", ax=ax)  
ax = dg.plot.scatter(0,1, marker='+', c='b',label = "target : +1", s=40, ax=ax)  
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
plt.xlabel('$x_1$', fontsize=16)
plt.ylabel('$x_2$', fontsize=16)
plt.show()
```
