## <span style='font-family:"Times New Roman"'> This code, generates data drawn from two circular symmetric multi-variate Gaussian distributions with specific means and variance. 

### <span style='font-family:"Times New Roman"'>Loading Required Modules:

```python
import matplotlib.pyplot as plt
import numpy as np
import math
plt.style.use('ggplot')
```

### <span style='font-family:"Times New Roman"'>Implemented Functions:  

```python
def circGuass(N,Me,Var):
    dimension = len(Me)
    a = np.array(Me)
    p = np.random.randn(dimension,N)
    h = np.tile(a, (2, N//2))
    x = math.sqrt(Var) * p + h
    return x
```

```python
M1 = float(input("Enter first set mean x coordinate (In this problem = 0) : "))
M2 = float(input("Enter first set mean y coordinate(In this problem = 0) : "))
M3 = float(input("Enter second set mean x coordinate(In this problem = 5) : "))
M4 = float(input("Enter second set mean y coordinate(In this problem = 5) : "))
Var = float(input("Enter variance of data set : (In this problem = 3)"))
N = int(input("Number of sample data : "))
```

```python
firstmean = [M1, M2]
secondmean = [M3, M4]
Me = firstmean;
X = circGuass(N, Me, Var)
rowonex= X[0]
rowtwox = X[1]
Me = secondmean
Y = circGuass(N, Me, Var)
rowoney= Y[0]
rowtwoy = Y[1]
```

```python
plt.figure(figsize=(8,8))
fig1 = plt.figure(1)
ax1 = fig1.gca()
plt.scatter(rowonex,rowtwox , color='blue', s=10)
plt.scatter(rowoney,rowtwoy , color='blue', s=10)
ax1.set_xlim(-5,10)
ax1.set_ylim(-5,10)
plt.xlabel('$x_1$', fontsize=16)
plt.ylabel('$x_2$', fontsize=16)
plt.show()
```
